# Resolvendo equações - Método de Newton

```julia
f(x) = x^2 - 2 # vamos achar o zero dessa função
fd(x) = 2x # precisamos entrar manualmente com a derivada
x = 1.0

# rodando o método 3 vezes já temos a precisão de 5 casas decimais
for i = 1:3
    x = x - f(x) / fd(x)
end
x, sqrt(2), f(x) 
```

```julia
function newton(f, fd, x; tolerância_absoluta = 1e-8, tolerância_relativa = 1e-8, max_tempo = 10.0, max_iterações = 1000)
    fx = f(x) # valor da função em x
    fdx = fd(x) # derivada no ponto x
    
    # critério de parada
    ϵ = tolerância_absoluta + tolerância_relativa * abs(fx) # quando o valor da função chega próximo de zero, conforme as tolerâncias passadas 
    
    iterações = 0
    # calcula quanto tempo o programa está levando
    t0 = time()
    Δt = time() - t0
    
    # testa os critérios de parada
    resolvido() = abs(fx) <= ϵ # é considerado resolvido quando o valor da função é menor ou igual a tolerância
    cansado() = (iterações >= max_iterações) || (Δt >= max_tempo)
    
    while !(resolvido() || cansado()) # roda enquanto nenhuma das condições de parada acima estão cumpridas    
        # devemos evitar o caso em que a derivada é muito próxima de zero
        # pois nesse caso, a próxima iteração será muito distante da atual
        # ou, caso seja zero, teremos o problema da divisão por zero
        if abs(fdx) <= ϵ
            exitflag = :derivada_nula
            return x, fx, iterações, exitflag, Δt
        end
        
        x = x - fx / fdx # o x da próxima iteração será o zero da reta tangente no x anterior
        fx = f(x)
        fdx = fd(x)
        
        iterações += 1
        Δt = time() - t0
    end
    
    exitflag = :desconhecido
    
    # verifica qual das condições de parada foram alcançadas para retornar um feedback
    if resolvido()
        exitflag = :sucesso
    else
        if iterações >= max_iterações
            exitflag = :max_iterações
        else
            exitflag = :max_tempo
        end
    end
    
    return x, fx, iterações, exitflag
end
```

```julia
newton(f, fd, 1.0) # fazendo o primeiro teste
# o resultado é equivalente a achar sqrt(2)
```

```julia
# testando com uma precisão ainda maior, nosso resultado fica com mais casas decimais corretas
newton(f, fd, big"1")
```

```julia
# fazendo o teste para mais uma função
f(x) = exp(x) * x - 1
fd(x) = exp(x) + x * exp(x)
newton(f, fd, 1.0)
```

```julia
using ForwardDiff

function eq_price(i)
    n = 48
    v = 40_000
    p = 1200

    if i == 0
        return p - v / n
    else
        return p - v * (1 + i)^n * i / ((1 + i)^n - 1)
    end
end

eq_price_derivada(i) = ForwardDiff.derivative(eq_price, i)

newton(eq_price, eq_price_derivada, 1.0)
```

#### Um exemplo que não converge bem

```julia
f(x) = exp(x + 5sin(x-10) -10) - 2 # uma função bem estranha
fd(x) = ForwardDiff.derivative(f, x)

newton(f, fd, 10.0) # converge se o valor dado estiver muito próximo da raiz
```

```julia
newton(f, fd, 5.0) # chega ao máximo de iterações
```

## Animação do método

```julia
using Plots

function mostra_animação(f, fd, a, b; xlim1 = -1.4, xlim2 = 1.4, ylim1 = -1.5, ylim2 = 1.5)
    anim = Animation()

    for x = a:0.01:b
        plot(f, -1.4, 1.4, leg=false)
        scatter!([x], [f(x)], c=:red, ms=2)
        
        local x
        for k = 1:10
            if abs(fd(x)) <= 1e-8
                continue
            end
            
            xp = x - f(x) / fd(x)
            plot!([x, xp, xp], [f(x), 0, f(xp)], c=:pink)
            scatter!([xp], [f(xp)], c=:pink, ms=2)
            
            x = xp
        end
        xlims!(xlim1, xlim2)
        ylims!(ylim1, ylim2)
        frame(anim)
    end
    
    gif(anim, "exemplo.gif", fps=10)
end
```

Segue uma ilustração da convergẽncia do método para cada um dos chutes no intervalo [-1.2, 1.2]


```julia
f(x) = x^3 - x
fd(x) = 3x^2 - 1

mostra_animação(f, fd, -1.2, 1.2)
```

O bloco abaixo mostra para qual dos zeros da função o método converge. Isso irá depender do chute inicial

Em vermelho converge para x = -1, em verde para x = 0, em azul para x = 1

```julia
X = []
C = []
colors = [:red, :green, :blue]

anim = Animation()

for x = -1.4:0.02:1.4
    plot(f, -1.4, 1.4, leg=false)
    scatter!([x], [f(x)], c=:red, ms=2)
 
    push!(X, x)
    
    local x
    for k = 1:3
        xp = x - f(x) / fd(x)
        plot!([x, xp, xp], [f(x), 0, f(xp)], c=:pink)
        scatter!([xp], [f(xp)], c=:pink, ms=2)

        x = xp
    end
    
    j = argmin(abs.(x .- [-1, 0, 1]))
   
    push!(C, colors[j])
    
    for (xi, ci) in zip(X, C)
        plot!([xi, xi], [-1.5, 1.5], c=ci)
    end
    
    xlims!(-1.4, 1.4)
    ylims!(-1.5, 1.5)
    frame(anim)
end

gif(anim, "exemplo.gif", fps=10)
```

## Análise da velocidade de convergência do método

O método possui convergência quadrática. Isso significa que a cada iteração, a quantidade de casas decimais corretas será dobrada. É um método extremamente rápido.

No gráfico é possivel ver que, a cada iteração, o erro decai quadraticamente. Ou seja, se na iteração 4 o erro era de 1e-4, na quinta iteração, o erro é 1e-8

```julia
# vamos plotar o erro do método

f(x) = x^2 - 2
fd(x) = 2x

N = collect(1:20)
Erro = [max(abs(newton(f, fd, 5.0, tolerância_relativa=1e-16, tolerância_absoluta=1e-16,max_iterações = n)[1] - sqrt(2.0)), 1e-16) / sqrt(2.0) for n in N]

scatter(N, Erro, yaxis=:log)
xlabel!("Número de iterações")
title!("Erro Relativo")
```
