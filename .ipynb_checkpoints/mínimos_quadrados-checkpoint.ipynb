```julia
# criando um conjunto de dados ruidosos para tentarmos achar uma função que modele o comportamento deles

n = 10
x = range(0, 1, length=n)

# aqui temos um modelo falso para testarmos o funcionamento do método
# o resultado final deve ser próximo de (3, 2)
y = 2x .+ 3 + randn(n) * 0.2 # randn retorna um número aleatório com probabilidade maior de estar ao redor de n

scatter(x, y, leg=false)
```

```julia
# criamos uma função erro, que é o somatório das diferenças entre os dados originais e o modelo
# o somatório está dividido por dois para simplificar o modelo, já que aparecerá um 2 multiplicando quando derivarmos
# a ideia é achar β₀ e β₁ que minimizem essa função erro

E(β₀, β₁) = sum((y[i] - β₀ - β₁ * x[i])^2 for i = 1:n) / 2
```

```julia
# aqui podemos fazer alguns chutes para β₀ e β₁, mas o ideal é fazer isso de forma automática

β₀ = 3.1
β₁ = 1.6

scatter(x, y, leg=false)
plot!(x -> β₀ + β₁ * x, extrema(x)..., c=:red, lw=2)
title!("Soma do Quadrados dos Resíduos = $(E(β₀, β₁))")
```

```julia
# aqui podemos ver o contorno e a superfície formada pela função erro. É um parabolóide eplíptico com concavidade para cima
# tive que deixar essa parte comentada pois estava quebrando o kernel

β₀range = range(2.5, 3.3, length=100)
β₁range = range(1.5, 2.5, length=100)

#p = contour(β₀range, β₁range, E, levels=20, layout=(1,2), size=(600,200))
#surface!(p[2], β₀range, β₁range, E)
```

```julia
# derivando a função erro em relação a β₀ e β₁
#teremos um sistema linear de duas equações, onde precisamos achar β₀ e β₁

M = [n  sum(x); sum(x)  sum(x .^ 2)]
c = [sum(y); sum(x .* y)]

# como podemos ver, a solução do sistema é realmente próxima do que foi modelado (3, 2)
βsol = M \ c
```

```julia
# podemos criar um modelo mais complexo que não seja polinômial
# nesse caso, queremos achar um modelo que seja uma combinação de senos e cossenos
# isso ainda é possível pois, por mais que as funções não sejam lineares, o sistema será linear em relação aos coeficientes

using Random
Random.seed!(0)

# gerando dados fictícios para teste
# a ideia do modelo é ver como se comporta determinado fenômeno ao longo de um ano
n = 365
t = range(0, 1, length=n)

# essas funções podem ser escolhidas da forma que quisermos. A escolha dependerá da natureza do problema
ϕ1(t) = sin(2π * t)
ϕ2(t) = cos(2π * t)
ϕ3(t) = 1.0

y = 6 * ϕ1.(t) + 4 * ϕ2.(t) + 12*ϕ3.(t) + randn(n)*1.1 +
    [rand() < 0.2 ? randn() * 3.5 : 0.0 for i = 1:n]

scatter(t, y, c=:lightblue, ms=3, leg=false)
```

```julia
# aqui, criamos manualmente a matriz que modela nosso sistema linear
# chegamos nela da mesma forma que a anterior, porém derivando a função erro em relação a β₀, β₁, β₂

M = zeros(3,3)
M[1,1] = sum(ϕ1(t[i]) * ϕ1(t[i]) for i = 1:n)
M[1,2] = sum(ϕ1(t[i]) * ϕ2(t[i]) for i = 1:n)
M[1,3] = sum(ϕ1(t[i]) * ϕ3(t[i]) for i = 1:n)
M[2,1] = M[1,2]
M[2,2] = sum(ϕ2(t[i]) * ϕ2(t[i]) for i = 1:n)
M[2,3] = sum(ϕ2(t[i]) * ϕ3(t[i]) for i = 1:n)
M[3,1] = M[1,3]
M[3,2] = M[2,3]
M[3,3] = sum(ϕ3(t[i]) * ϕ3(t[i]) for i = 1:n)

c = zeros(3)
c[1] = sum(y[i] * ϕ1(t[i]) for i = 1:n)
c[2] = sum(y[i] * ϕ2(t[i]) for i = 1:n)
c[3] = sum(y[i] * ϕ3(t[i]) for i = 1:n)

# a solução deverá ser próxima do modelo que criamos anteriormente, ou seja, (6, 4, 12)
β = M \ c
```

```julia
# aqui podemos ver como, de fato a solução encontrada modela bem o comportamento geral dos dados

scatter(t, y, c=:lightblue, ms=3, leg=false)
plot!(t -> β[1] * ϕ1(t) + β[2] * ϕ2(t) + β[3] * ϕ3(t),
        extrema(t)..., c=:red, lw=2)
```

```julia
# podemos generalizar essa ideia, criando uma função que recebe os pontos e as funções que serão usadas para modelar

function mínimos_quadrados(x, y, F)
    quantidade_pontos = length(x)
    quantidade_funções = length(F)
    
    M = zeros(quantidade_funções, quantidade_funções)
    c = zeros(quantidade_funções)
    
    # aqui criamos as matrizes que modelam o sistema linear que devemos resolver
    for j = 1:quantidade_funções
        for k = 1:quantidade_funções
            M[j,k] = sum(F[j](x[i]) * F[k](x[i]) for i = 1:quantidade_pontos)
        end
        c[j] = sum(y[i] * F[j](x[i]) for i = 1:quantidade_pontos)
    end
    
    β = M \ c # solução do sistema
    h(x) = sum(F[j](x) * β[j] for j = 1:quantidade_funções) # função final que modela o comportamento dos dados
    
    return β, h
end
```

```julia
# testando nossa implementação

using LinearAlgebra, Random

Random.seed!(0) # fixamos uma semente pro resultado ser sempre o mesmo

quantidade_pontos = 20

x = range(-1, 1, length=quantidade_pontos)
y = 1 ./ (1 .+ 15 * x.^2) + randn(quantidade_pontos) * 0.1

modelo_real(x) = 1 ./ (1 .+ 15 * x.^2) # modelo final onde queremos chegar

scatter(x, y, leg=false)

# podemos testar várias combinações de funções diferentes e ver como elas modelam os dados

# F = [x -> 1.0, x -> x, x -> x^2, x -> x^3, x -> x^4, x -> x^5,
#     x -> x^6, x -> x^7, x -> x^8, x -> x^10, x -> x^12, x -> x^14,
#     x -> x^16, x -> x^18, x -> x^20]

# F = [x -> x^j for j = 0:14]

# F = [x -> exp(-k * x^2) for k = 1:20]

F = [x -> 1 / (1 + 15x^2)]

β, modelo_resultado = mínimos_quadrados(x, y, F)
r = y - modelo_resultado.(x)
SQR = norm(r)^2

plot!(modelo_resultado, extrema(x)..., c=:red, lw=2)
plot!(modelo_real, extrema(x)..., c=:blue)
title!("SQR = $SQR")
```

```julia
# um modelo que não aproxima bem, o erro é bem alto

F = [x -> cos(π * x), x -> sin(π * x)]

β, modelo_resultado = mínimos_quadrados(x, y, F)
r = y - modelo_resultado.(x)
SQR = norm(r)^2

plot!(modelo_resultado, extrema(x)..., c=:red, lw=2)
plot!(modelo_real, extrema(x)..., c=:blue)
title!("SQR = $SQR")
```

# Exercício 2


2) A polícia chega ao local de um assassinato às 15h. Eles imediatamente medem e registram a temperatura do corpo, que é 34 graus Celsius, e inspecionam minuciosamente a área. Quando terminam a inspeção, são 16:30h. Eles medem novamente a temperatura do corpo, que caiu para 30 graus Celsius. Eles esperam mais 1 hora, e medem a temperatura de novo, que caiu para 25 Graus Celsius. A temperatura na cena do crime permaneceu estável em 20 graus Celsius e a temperatura normal do corpo é 37 graus Celsius.   Sabendo que a temperatura respeita a Lei de resfriamento de Newton (http://www.if.ufrgs.br/tex/fis01043/20011/Adriano/intro.html) , use mínimos quadrados não linear (como fizemos na aula 12) para descobrir que horas a pessoa foi assassinada.

Considerando a Lei de resfriamento de Newton, podemos modelar os dados com a seguinte função
$$y = \beta _{1} \cdot e^{\beta_{2} \cdot t} + \beta_{3} $$

Para usar o método dos mínimos quadrados, podemos linearizar esse modelo tirando o log dos ambos lados da equação

$$ln(y - \beta_{3}) = ln(\beta _{1} \cdot e^{\beta_{2} \cdot t}) $$
$$ln(y - \beta_{3}) = ln(\beta _{1}) + ln(e^{\beta_{2} \cdot t}) $$
$$ln(y - \beta_{3}) = ln(\beta _{1}) + \beta_{2} \cdot t $$

Então podemos aplicar o método na seguinte equação
$$\bar{y} = \bar{\beta _{1}} + \bar{\beta_{2}} \cdot t $$

Onde:

$ y = e^{\bar{y}} +  \beta_{3}$

$ \beta _{1} = e^{\bar{\beta _{1}}}$

$ \beta _{2} = \beta _{2}$

Abaixo aplicamos o método nos dados fornecidos.

```julia
using LinearAlgebra, Plots

# range para mostrar o gráfico
x = range(14, 20, length=20)

# dados do tempo em que cada medição foi feita
t = [15, 16.5, 17.5]

# a temperatura medida em cada horário
y = [34, 30, 25]

# as medições foram transformadas para encaixarem no modelo linearizado
# o β₃ foi chutado como 20 pois de acordo com a lei, seria a temperatura final
β₃ = 20
y2 = [log(34 - β₃), log(30 - β₃), log(25 - β₃)]

# as funções usadas no modelo
F = [t -> 1.0, t -> t]

# rodando os mínimos quadrados
β, modelo_resultado = mínimos_quadrados(t, y2, F)
println(β)

# precisamos desfazer a transformação feita acima para obter a função orignal que queríamos
modelo_transformado(t) = exp(β[1]) * exp(β[2]*t) + β₃

# calculando o erro em relação às medições iniciais
r = y - modelo_transformado.(t)
SQR = norm(r)^2

# plotando a função e os pontos medidos
plot(modelo_transformado, extrema(x)..., c=:blue)
scatter!(t, y, c=:lightblue, ms=3, leg=false)
title!("SQR = $SQR")
```

Usando o modelo encontrado, para achar o horário em que a pessoa faleceu, devemos achar o valor de $t$ quando a temperatura é 37.

Para isso, podemos usar o método da bisseção para resolver a equação $modelo\_transformado(t) = 37$

```julia
function bisseção(f, a, b; tolerância_absoluta = 1e-8, tolerância_relativa = 1e-8, max_tempo = 10.0, max_iterações = 1000)
    fa = f(a)
    fb = f(b)
    
    # critérios de parada
    ϵ = tolerância_absoluta + tolerância_relativa * max(abs(fa), abs(fb)) # quando o valor da função chega próximo de zero, conforme as tolerâncias passadas 
    ϵ_intervalo = tolerância_absoluta + tolerância_relativa * abs(b - a) # quando o tamanho do intervalo se tornar muito curto
    
    # testes iniciais de parada
    if abs(fa) <= ϵ
        return a, fa, :sucesso # se algum dos valores passados já for o zero da função
    elseif abs(fb) <= ϵ
        return b, fb, :sucesso
    elseif fa * fb >= 0
        return a, fa, :falha_sinais_opostos # se o intervalo passado for ruim e não tiver variação de sinal
    end
    
    ponto_médio = (a + b) / 2
    fm = f(ponto_médio)
    iterações = 0
    
    # calcula quanto tempo o programa está levando
    t0 = time()
    Δt = time() - t0
    
    # testa os critérios de parada
    resolvido() = (abs(fm) <= ϵ) || (abs(b - a) <= ϵ_intervalo)
    cansado() = (iterações >= max_iterações) || (Δt >= max_tempo)
    
    while !(resolvido() || cansado()) # roda enquanto nenhuma das condições de parada acima estão cumpridas
        if fa * fm < 0 # se os sinais de fa e fm forem opostos, reduzimos o intervalo pelo lado b
            b = ponto_médio
            fb = fm
        else
            a = ponto_médio
            fa = fm
        end
        
        ponto_médio = (a + b) / 2
        fm = f(ponto_médio)
        
        iterações += 1
        Δt = time() - t0
    end
    
    exitflag = :desconhecido
    
    # verifica qual das condições de parada foram alcançadas para retornar um feedback
    if resolvido()
        exitflag = :sucesso
    else
        if iterações >= max_iterações
            exitflag = :max_iterações
        else
            exitflag = :max_tempo
        end
    end
    
    return ponto_médio, fm, iterações, exitflag
end
```

```julia
# a função a ser usada no método da bisseção
f(t) = modelo_transformado(t) - 37

# de acordo com o gráfico acima, o zero parece estar entre 14.2 e 15
bisseção(f, 14.2, 15.0, tolerância_absoluta = 1e-14, tolerância_relativa = 1e-14)
```
